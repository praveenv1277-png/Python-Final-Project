##  EDA and Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/praveenv1277-png/Python-Final-Project/refs/heads/main/Imports%20at%20Principal%20Commodity%20Level_Data%20(1).csv")

df.head()
```

```python
df.columns = [
    'date',
    'state_name',
    'state_code',
    'country',
    'port',
    'commodity_code',
    'commodity',
    'units',
    'quantity',
    'dollars_value',
    'inr_value'
]

print(df.columns)
```

# Quantity Distribution


How are import quantities distributed across import transactions?

```python
plt.figure(figsize=(8,5))
sns.histplot(df['quantity'], bins=20, kde=True)
plt.title("Distribution of Import Quantity")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

# Key Insights

1.The dataset contains 500 import transactions with shipment quantities ranging
from 0 to 8,446,150 units.

2.A large proportion of transactions have very low or zero recorded quantity, while only a few transactions contain very large shipment volumes.

3.This indicates that the dataset is dominated by small-volume imports with a limited number of bulk shipments.

4.The distribution is highly right-skewed, showing that bulk imports are relatively uncommon.

5.Businesses should manage regular shipments efficiently while planning separately for occasional bulk imports.



# USD Import Value Distribution


How are import values distributed in USD?

```python
plt.figure(figsize=(8,5))
sns.histplot(df['dollars_value'], bins=20, kde=True,color='green')
plt.title("Distribution of Import Value (USD)")
plt.xlabel("USD Value")
plt.ylabel("Frequency")
plt.show()
```

# Key Insights

1.The import value ranges from USD 0 to USD 83.49 million, showing a very wide variation in transaction values.

2.Most transactions are concentrated in the lower USD value range, while only a limited number exceed USD 10 million.

3.The distribution confirms that a few high-value imports contribute a major share of the total import expenditure.

4.The large spread indicates substantial price differences between imported commodities.

5.Financial monitoring should focus on high-value transactions because they have the greatest impact on total import spending.

 # Top 10 States by Import Transactions
 Which Indian states record the highest number of import transactions, indicating major import activity across the country?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    y='state_name',
    order=df['state_name'].value_counts().head(10).index
)

plt.title("Top 10 States Based on Import Transactions")
plt.xlabel("Number of Transactions")
plt.ylabel("State")
plt.show()
```

# Key Insights

1.The chart identifies the top 10 Indian states with the highest number of import transactions, highlighting the regions where import activities are most concentrated.

2.The top-ranked state recorded the highest number of import transactions, indicating it is one of the country's most active import centers.

3.The difference in transaction counts among the top 10 states suggests that import operations are concentrated in a few key states rather than being evenly distributed across India.

4.States appearing in the top 10 are likely to have major ports, airports, industrial zones, or logistics infrastructure that support international trade.

5.A higher number of transactions does not necessarily indicate a higher import value; it reflects the frequency of import activities. Some states may have fewer transactions but much higher import values.

6.This analysis helps businesses identify high-activity import regions, enabling better decisions for warehouse placement, transportation planning, and supply chain expansion.

# Top 10 Imported Commodities

Which commodities are imported most frequently?

```python
top=df['commodity'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top.values,y=top.index)

plt.title("Top 10 Imported Commodities")
plt.xlabel("Number of Transactions")
plt.ylabel("Commodity")
plt.show()
```

# Key Insights

1.Iron And Steel is the most frequently imported commodity with 20 transactions (4.0% of all records).

2.Paper, Paper Board And Product ranks second with 19 transactions (3.8%).
Electric Machinery And Equipme ranks third with 16 transactions (3.2%).
These three commodity groups together account for 55 import transactions (11% of the dataset).

3.The results indicate consistent demand for industrial raw materials and manufacturing-related products.

# Quantity vs USD Value

Is there a relationship between import quantity and import value?

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,
                x='quantity',
                y='dollars_value')

plt.title("Quantity vs USD Value")
plt.show()
```

# Key Insights

1.The correlation between Quantity and USD Value is only 0.0969, indicating an extremely weak relationship.

2.Large shipment quantities do not necessarily correspond to expensive imports.
Several transactions with relatively small quantities have very high USD values, suggesting imports of premium-priced commodities.

3.The scatter plot confirms that commodity price has a greater influence on import value than shipment volume.
Procurement decisions should therefore consider both quantity and unit value.

 # Country-wise Import Value

Which countries contribute the highest import value?

```python
country=df.groupby('country')['inr_value'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=country.values,y=country.index)

plt.title("Top 10 Countries by Import Value")
plt.xlabel("Total INR Value")
plt.ylabel("Country")
plt.show()
```

# Key Insights

1.USA is the largest supplier with approximately ₹13.11 billion in import value.
Switzerland contributes approximately ₹11.63 billion, making it the second-largest supplier.

2.China P Rp contributes approximately ₹4.06 billion, ranking third.

3.The combined import value of the top three supplier countries exceeds ₹28.8 billion, indicating a strong dependence on a limited number of trading partners.
Businesses should monitor supplier concentration risk and evaluate opportunities for supplier diversification.

# State-wise Import Value

Which Indian states record the highest import value?

```python
state=df.groupby('state_name')['inr_value'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=state.values,y=state.index)

plt.title("Top States by Import Value")
plt.xlabel("Total INR Value")
plt.ylabel("State")
plt.show()
```

# Key Insights

1.Gujarat records the highest import value at approximately ₹23.92 billion, making it the leading import hub.

2.Delhi follows with approximately ₹12.53 billion, while Maharashtra records ₹7.06 billion.

3.The combined import value of these three states exceeds ₹43 billion, representing a substantial share of total imports.

4.These states serve as major logistics and industrial centers for import activities.

5.Investments in warehousing and transportation infrastructure should primarily focus on these high-import regions.

# Correlation Heatmap


How are quantity, USD value, and INR value related?

```python
plt.figure(figsize=(6,5))

sns.heatmap(df[['quantity',
                'dollars_value',
                'inr_value']].corr(),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()
```

# Key Insights

1.USD Value and INR Value have a correlation coefficient of 0.9994, indicating an almost perfect positive relationship.

2.Quantity and USD Value have a correlation of only 0.0969, showing that shipment size has little influence on financial value.

3.Quantity and INR Value have a correlation of 0.0900, confirming a similarly weak relationship.

4.Import value is primarily determined by commodity price rather than shipment volume.

5.Businesses should evaluate procurement performance using monetary value instead of quantity alone.

# Commodity vs Country Heatmap


Which countries are the primary suppliers for different commodities?

```python
pivot=pd.pivot_table(df,
                     values='inr_value',
                     index='commodity',
                     columns='country',
                     aggfunc='sum',
                     fill_value=0)

plt.figure(figsize=(14,8))
sns.heatmap(pivot,cmap='YlGnBu')

plt.title("Commodity vs Country Import Value")
plt.show()
```

# Key Insights

1.The heatmap shows that supplier contributions vary significantly across commodity categories.

2.Industrial commodities such as Iron And Steel and Electric Machinery are sourced from multiple countries, reducing dependence on a single supplier.

3.Several commodity-country combinations contribute substantially higher import values than others, indicating specialized sourcing patterns.

4.The visualization highlights strategic supplier relationships that are critical for procurement planning.

5.Businesses can use this information to strengthen supplier diversification and reduce supply chain risks.
